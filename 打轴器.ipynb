## 输出帧图像测试

```python
import os
import sys
import time
import cv2
import argparse
 
 
def process_video(input, output):
    if not os.path.exists(output):
        os.makedirs(output)
 
    vid = cv2.VideoCapture(input)
    proc_frames = 0
    ret, frame = vid.read()
    while ret:
        out_file = output + '/{:0>6d}.jpg'.format(proc_frames)
        cv2.imwrite(out_file, frame)
        ret, frame = vid.read()
        proc_frames += 1
    vid.release()
 
 
if __name__ == "__main__":
 
    input = "./白子Momotalk-1.mp4"
    output="images"
    start_time = time.time()
    process_video(input, output)
 
    print(
        "Method {}: Input:{}, Time taken: {}".format(
            sys.argv[0], input, time.time() - start_time
        )
    )
```

```python
def check(img1, img2=None):
    if(img2 is None):
        if (img1 ** 2).sum() / img1.size * 100 > 1:
            return True
    else:
        if ((img2 - img1) ** 2).sum() / img1.size * 100 > 1:
            return True
    return False
```

```python
# 导入easyocr
import easyocr
# 创建reader对象
reader = easyocr.Reader(['ja']) 

video_filename = './白子Momotalk-1.mp4'
videoCap = cv2.VideoCapture(video_filename)

fps = videoCap.get(cv2.CAP_PROP_FPS) # 帧频
total_frames = int(videoCap.get(cv2.CAP_PROP_FRAME_COUNT)) # 视频总帧数
image_size = (int(videoCap.get(cv2.CAP_PROP_FRAME_HEIGHT)), int(videoCap.get(cv2.CAP_PROP_FRAME_WIDTH))) # 图像尺寸

cnt = 0
ret, last_frame = videoCap.read()
start, end = 0, 0

'''
对白：有名字
名字出现-文本消失-名字改变-文本消失(名字消失)
名字出现、名字改变、文本消失判断时轴开始
名字出现(上一帧)、名字改变(上一帧)、文本消失(上一帧)判断时轴结束

旁白：无名字
文本出现-文本消失
'''

while True:
    ret, frame = videoCap.read()
    cnt += 1
    if not ret:
        print('video: %s finish at %dth frame.' % (video_filename, cnt))
        break
    if check(nameImg(last_frame), nameImg(frame)) and not check(nameImg(last_frame)) and check(nameImg(frame)): #名字出现
        end = cnt
        print("name app end", cnt-1)
        if end-start>15 and start != 0:
            result = reader.readtext(last_frame[765:1060,:,:])
            res = ""
            for i in result:
                res = res+" "+i[1]
            print("%d - %d: %s" % (start, end-1, res))
            # showImg(last_frame)
        print("name app start", cnt)
        start = cnt
    elif check(textImg(last_frame), textImg(frame)) and check(textImg(last_frame)) and not check(textImg(frame)): #文本消失
        end = cnt
        print("text dis end", cnt-1)
        if end-start>15 and start != 0:
            result = reader.readtext(last_frame[765:1060,:,:])
            res = ""
            for i in result:
                res = res+" "+i[1]
            print("%d - %d: %s" % (start, end-1, res))
            # showImg(last_frame)
        print("text dis start", cnt)
        start = cnt
    elif check(nameImg(last_frame), nameImg(frame)) and check(nameImg(last_frame)) and check(nameImg(frame)): #名字改变
        end = cnt
        print("name change end", cnt-1)
        if end-start>15 and start != 0:
            result = reader.readtext(last_frame[765:1060,:,:])
            res = ""
            for i in result:
                res = res+" "+i[1]
            print("%d - %d: %s" % (start, end-1, res))
            # showImg(last_frame)
        print("name change start", cnt)
        start = cnt
    elif check(textImg(last_frame), textImg(frame)) and not check(textImg(last_frame)) and check(textImg(frame)): #文本出现
        if cnt-3 != start and cnt-4 != start:
            print("text app", cnt)
            start = cnt
    last_frame = frame
```
