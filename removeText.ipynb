```python
import cv2
import matplotlib.pyplot as plt 
```

```python
def show(src):
    t = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
    plt.imshow(t)
    plt.show()
```

```python
def Inpainting(srcImg,x,y,w,h,type=0):
    src = srcImg[x:x+w,y:y+h].copy()
    gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY,1)
    if type == 0: #白字
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
    elif type == 1:
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    maskImg = cv2.dilate(thresh, kernel)
    inpaintImg = cv2.inpaint(src,maskImg,3,cv2.INPAINT_NS)
    srcImg[x:x+w,y:y+h] = inpaintImg
    show(srcImg)
```

```python
title = cv2.imread("title.jpg")
show(title)
Inpainting(title,450,400,180,600,1)
```

```python
dialogue = cv2.imread("dialogue.jpg")
show(dialogue)
Inpainting(dialogue,880,120,100,1050)
```

```python
choice = cv2.imread("choice.jpg")
show(choice)
Inpainting(choice,400,520,50,450,1)
```

```python
name_school = cv2.imread("name_school.jpg")
show(name_school)
Inpainting(name_school,820,80,60,400)
Inpainting(name_school,880,100,100,1050)
```
